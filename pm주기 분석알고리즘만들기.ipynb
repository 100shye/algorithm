## reset index 임의로 만들기

```python
import pandas as pd
import numpy as np

df_total = pd.DataFrame()

for num in range(4): # pm을 4번 진행
    df= pd.DataFrame()
    size = np.random.randint(70, 100, 1)[0]
    i = np.random.randint(0, 20, 1)[0]
    
    # pm 인덱스는 gks
    df['pm_idx'] = np.random.randint(i*20+0, i*20+100, size)
    
    # sig1은 pm할때마다 랜덤으로 값이 조금씩 변하고 암캐나 흔들림
    df['sig1'] = np.random.randint(i*20+0, i*20+10, size)
    
    df = df.sort_values(by='pm_idx')    
    # sig2 는 pm 할때마다 시작점은 랜덤이지만 약간씩 증가하는경향임
    
    df_total = pd.concat([df_total, df])
df_total = df_total.reset_index(drop=True)
```
