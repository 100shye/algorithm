# DOE(Design Of Experiment, 실험계획법)
 - 효율적인 실험 방법을 설계하고 결과제를 제대로 분석하는 것을 목표로 하는 실험 분야
 - 실험의 독립변수 X, 종속변수를 y라고 하자
 - DOE를 진행하기 전에 두가지를 정해야 한다. <br>
   1) Y관한 어떤 모델을 만들고 싶은가(ex number of polynomial terms) <br>
   2) x값의 범위, level <br>
 - 실험계획법은 크게 두가지로 나눌 수 있다. <br>
   1) Screening DOE: 초기 주요 영향변수를 알아내기 위한 실험계획법 <br>
   2) Augment DOE: 추가적인 실험을 진행하기 위한 실험계획법

##### DOE 결과 해석
 - 독립변수 X가 p, 반응 y가 N개 존재하면 N개의 회귀식이 생긴다.

# Screening DOE
 - 실험 초기에 여러 변수들이 결과에 미치는 영향력을 대략적으로 파악해보고 싶을 때 사용하는 실험계획법
 - 부분적인 실험을 진행하니까 fractional factorial DOE라고도 불린다.
 - 대표적인 Screening DOE 방식으로는 PBD(Plackett Burman Design)이 있다.
 - Screening DOE의 장단점은 아래와 같다.
  - 장점: 진행 할 실험수를 줄이는 것이 가능함.
  - 단점: 실험수를 줄이니까 인자간의 상관성도 불명확해지고 반응에 미치는 영향성 정보가 불완전하다.
 - 단점을 보안하기 위해 Screening DOE로 주 영향을 파악한 뒤에는 목표에 따라 추가실험을 진행한다.
 
 
 ref: https://www.americanpharmaceuticalreview.com/Featured-Articles/348755-How-to-Utilize-Design-of-Experiment-DoE-Principles-for-Developing-Robust-Analytical-Methods-for-QC-Environments/

# Screening DOE - Plackett-Burman Design
 - 대표적인 최소 실험 디자인 중의 하나이다. 주 영향인자를 스크리닝하는 것이 가능함.
 - 4-7개 factor 있으면 8회의 실험, 8-11의 factor있으면 12회 실험~~ 이렇게 4의 배수 실험이 필요함.

 ref: https://www.theopeneducator.com/doe/Fractional-Factorial-Design-of-Experiments/Plackett-Burman

# Python으로 Screening DOE 만들기
 - doepy package 설치 
 - ref: https://doepy.readthedocs.io/en/latest/

```python
from doepy import build
df = build.plackett_burman(
{'Pressure':[40, 70],
'Temperature':[290,  350],
'Flow rate':[0.2,0.4],
'Time':[5,11]},
)
df
```

#### 4개 factor 기준 PB Design 설계 시 8개 실험이 생성 되었다.
